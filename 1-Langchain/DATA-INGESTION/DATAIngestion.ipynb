```python
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader = WebBaseLoader(web_paths=("https://en.wikipedia.org/wiki/Google",),
bs_kwargs=dict(parse_only = bs4.SoupStrainer(
    class_ = ("mw-page-title-main")
)))
```

```python
loader.load()
```
